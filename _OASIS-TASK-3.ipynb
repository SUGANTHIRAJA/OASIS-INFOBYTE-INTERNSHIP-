```python
X=CPP[['carlength','carwidth','carheight','curbweight','enginesize','boreratio','car_ID','symboling','wheelbase','compressionratio','horsepower','peakrpm','citympg','highwaympg']]
y=CPP['price']
```

```python
X.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=10)
X_train.shape,y_train.shape
```

```python
X_test.shape,y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
#create the instance of linear regression
model= LinearRegression()
```

```python
#fitting the model to the dataset
model.fit(X_train,y_train)
```

```python
y_predict=model.predict(X_test)
y_predict
```

```python
from sklearn.metrics import mean_squared_error,r2_score
mse=mean_squared_error(y_test,y_predict)
r2 = r2_score(y_test, y_predict)
```

```python
print("Mean Squared Error:", mse)
print("R-squared Score:", r2)
```

```python
from sklearn import metrics

print('MAE:',metrics.mean_absolute_error(y_predict,y_test))
print('RMSE:',np.sqrt(metrics.mean_squared_error(y_predict,y_test)))
print('R-Squared',metrics.r2_score(y_predict,y_test))
```
